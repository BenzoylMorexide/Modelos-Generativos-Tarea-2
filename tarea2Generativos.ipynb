```python
import os
os.makedirs("imagenes", exist_ok=True)

prompts = {
    "musica_corto":   "A wooden cello.",
    "musica_medio":   "A wooden cello resting on a stage under a spotlight.",
    "musica_largo":   "A highly detailed vintage wooden cello resting on a grand concert hall stage, illuminated by a dramatic warm spotlight, cinematic lighting, 8k resolution.",
    "cafe_corto":     "A cup of coffee.",
    "cafe_medio":     "A cup of specialty pour-over coffee on a wooden table.",
    "cafe_largo":     "A glass carafe of specialty pour-over coffee pouring into a wooden mug, golden ratio, cozy cafe setting, afternoon sunlight, photorealistic.",
    "fantasia_corto": "A fantasy knight.",
    "fantasia_medio": "A fantasy knight with a glowing sword in a dark forest.",
    "fantasia_largo": "A heavily armored fantasy knight holding a glowing magical sword, standing in a dense magical forest, fog, highly detailed, open world RPG concept art, digital painting.",
    "espacio_corto":  "A spaceship.",
    "espacio_medio":  "A metallic spaceship landing on a strange alien planet.",
    "espacio_largo":  "A futuristic almighty spaceship exploring a vibrant alien planet with vibrant jungles and multiple moons in the sky, high quality, masterpiece.",
    "libro_corto":  "An open book.",
    "libro_medio":  "An open book with pages seemingly old written in an old language.",
    "libro_largo":  "An ancient, thick open book resting on a dusty wooden desk in a dimly lit library, its yellowed parchment pages filled with classical German literary text in ornate Gothic script, illuminated by a warm flickering candle, macro photography, highly detailed, 8k resolution.",
}

adversariales = {
    "adv_contradiccion": "A silent loud explosion in a bright dark room.",
    "adv_abstracto":     "The feeling of nostalgia on a Tuesday afternoon.",
    "adv_espacial":      "A cat sitting above a dog that is below a bird that is inside a box.",
}

semillas = [42, 123, 999]

def generar(prompt, seed, nombre):
    generador = torch.Generator(device="cuda").manual_seed(seed)
    t = time.time()
    resultado = pipeline(
        prompt=prompt,
        generator=generador,
        num_inference_steps=4,
        guidance_scale=0.0
    )
    imagen = resultado.images[0]
    ruta = f"imagenes/{nombre}_seed{seed}.png"
    imagen.save(ruta)
    print(f" [{contador}/{total}] {ruta} — {time.time()-t:.1f}s")
    return ruta

total = len(prompts) * len(semillas) + len(adversariales)
contador = 0

for nombre, prompt in prompts.items():
    for seed in semillas:
        contador += 1
        generar(prompt, seed, nombre)

for nombre, prompt in adversariales.items():
    contador += 1
    generar(prompt, seed=42, nombre=nombre)

print(f"\nListo, {total} imágenes en imagenes/")
```

```python
import os
import pandas as pd
from PIL import Image
import numpy as np
from brisque import BRISQUE
from transformers import CLIPProcessor, CLIPModel

brisque_obj = BRISQUE(url=False)
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch16")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch16")
# se declaran de nuevo
prompts_dict = {
    "musica_corto":   "A wooden cello.",
    "musica_medio":   "A wooden cello resting on a stage under a spotlight.",
    "musica_largo":   "A highly detailed vintage wooden cello resting on a grand concert hall stage, illuminated by a dramatic warm spotlight, cinematic lighting, 8k resolution.",
    "cafe_corto":     "A cup of coffee.",
    "cafe_medio":     "A cup of specialty pour-over coffee on a wooden table.",
    "cafe_largo":     "A glass carafe of specialty pour-over coffee pouring into a wooden mug, golden ratio, cozy cafe setting, afternoon sunlight, photorealistic.",
    "fantasia_corto": "A fantasy knight.",
    "fantasia_medio": "A fantasy knight with a glowing sword in a dark forest.",
    "fantasia_largo": "A heavily armored fantasy knight holding a glowing magical sword, standing in a dense magical forest, fog, highly detailed, open world RPG concept art, digital painting.",
    "espacio_corto":  "A spaceship.",
    "espacio_medio":  "A metallic spaceship landing on a strange alien planet.",
    "espacio_largo":  "A futuristic almighty spaceship exploring a vibrant alien planet with vibrant jungles and multiple moons in the sky, high quality, masterpiece.",
    "libro_corto":    "An open book.",
    "libro_medio":    "An open book with pages seemingly old written in an old language.",
    "libro_largo":    "An ancient, thick open book resting on a dusty wooden desk in a dimly lit library, its yellowed parchment pages filled with classical German literary text in ornate Gothic script, illuminated by a warm flickering candle, macro photography, highly detailed, 8k resolution.",
    "adv_contradiccion": "A silent loud explosion in a bright dark room.",
    "adv_abstracto":     "The feeling of nostalgia on a Tuesday afternoon.",
    "adv_espacial":      "A cat sitting above a dog that is below a bird that is inside a box."
}

resultados = []
archivos = [f for f in os.listdir("imagenes") if f.endswith(".png")]

print(f"\nEvaluando {len(archivos)} imágenes...")

for i, archivo in enumerate(archivos, 1):
    ruta = os.path.join("imagenes", archivo)

    nombre_base = archivo.split("_seed")[0]
    prompt_texto = prompts_dict.get(nombre_base, "")

    img_pil = Image.open(ruta).convert("RGB")

    img_np = np.asarray(img_pil)
    score_brisque = brisque_obj.score(img=img_np)

    inputs = clip_processor(text=[prompt_texto], images=img_pil, return_tensors="pt", padding=True)
    outputs = clip_model(**inputs)
    score_clip = outputs.logits_per_image.item()

    print(f"[{i}/{len(archivos)}] {archivo} | BRISQUE: {score_brisque:.2f} | CLIP: {score_clip:.2f}")

    resultados.append({
        "archivo": archivo,
        "prompt": prompt_texto,
        "brisque": score_brisque,
        "clip": score_clip
    })


df = pd.DataFrame(resultados)
df.to_csv("metricas.csv", index=False)
```
